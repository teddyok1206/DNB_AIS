```python
import pandas as pd
import os
```

```python
# 기본 정보 확인 (1)

raw_csv_list = os.listdir("./toTW")
raw_csv_list.sort()
print(raw_csv_list)
```

```python
# 기본 정보 확인 (2)

csv_dict = dict()
cnt = 0

for csv_name in raw_csv_list:
    if len(csv_name) != 43:
        print(len(csv_name))
    parts = csv_name.split("_")
    date = parts[1]
    time = parts[2]
    if date not in csv_dict.keys():
        csv_dict[date] = [time]
    else:
        csv_dict[date].append(time)
for key in csv_dict.keys():
    if (len(csv_dict[key]) == 2):
        cnt += 1
    else:
        print(key, csv_dict[key])
print(cnt)
print(len(raw_csv_list))
```

```python
# 1600 ~ 1700 데이터와 1700 ~ 1800 데이터를 하나의 csv 파일로 합치기 (걸쳐 있는 위성 영상이 있음)

cnt = 0
for i in range(227, len(raw_csv_list), 2):
    if ((i + 1) >= len(raw_csv_list)):
        break
    csv_1600 = raw_csv_list[i]
    csv_1700 = raw_csv_list[i + 1]
    parts_1600 = csv_1600.split("_")
    parts_1700 = csv_1700.split("_")
    same_date = (parts_1600[1] == parts_1700[1])
    exact_time = ((parts_1600[2] == "160000") and (parts_1700[2] == "170000"))
    if (same_date and exact_time):
        df_1600 = pd.read_csv("./toTW/"+csv_1600)
        df_1700 = pd.read_csv("./toTW/"+csv_1700)
        df_merged = pd.concat([df_1600, df_1700], ignore_index=True)
        fileName = "./toTW/AIS/" + ("_".join([parts_1600[0], parts_1600[1], "160000-180000"])) + ".csv"
        df_merged.to_csv(fileName, index=False)
        cnt += 1

print("result cnt :", cnt)
```
