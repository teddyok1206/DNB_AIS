```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
from matplotlib.patches import Patch
import numpy as np
from shapely import wkt
from sklearn.cluster import KMeans
import cartopy.crs as ccrs
import cartopy.feature as cfeature

df = pd.read_csv("./pixel_overlap-allITPL.csv")
print(df.head())
```

```python
import numpy as np
import matplotlib.pyplot as plt

# (Date, Time)별 overlap pixel 개수
pixel_count_per_dt = (
    df.groupby(["Date", "Time"])
      .size()
)

# 0 제거 (로그 스케일 필수)
vals = pixel_count_per_dt[pixel_count_per_dt > 0]

# 로그 간격 bin 생성
bins = np.logspace(
    np.log10(vals.min()),
    np.log10(vals.max()),
    50
)

plt.figure(figsize=(7,4.5))
plt.hist(
    vals,
    bins=bins,
    edgecolor="black",
    linewidth=0.6,
    color="#4C72B0"
)

plt.xscale("log")
plt.yscale("log")   # ★ 이게 깔쌈함의 핵심 중 하나

plt.xlabel("Number of overlap pixels per image (log scale)")
plt.ylabel("Frequency (log scale)")
plt.title("Distribution of overlap pixel count per Image\n(S-NPP, JPSS-1, JPSS-2)")

plt.grid(which="both", axis="both", alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
print(pixel_count_per_dt.min())
print(pixel_count_per_dt.max())
print(pixel_count_per_dt.mean())
print(pixel_count_per_dt.std())
```

```python
# (Date, Time)별 count 평균
mean_count_per_dt = (
    df.groupby(["Date", "Time"])["count"]
      .mean()
)

# 0 제거 (논리적으로도 맞음)
vals = mean_count_per_dt[mean_count_per_dt > 0]

import numpy as np

# 로그 bin 생성
bins = np.logspace(
    np.log10(vals.min()),
    np.log10(vals.max()),
    50
)

plt.figure(figsize=(7,4.5))
plt.hist(
    vals,
    bins=bins,
    edgecolor="black",
    linewidth=0.6,
    color="#4C72B0"
)

plt.xscale("log")
plt.yscale("log")   # ← 이게 핵심 (선택이지만 강력 추천)

plt.xlabel("Mean count per overlap pixel (log scale)")
plt.ylabel("Frequency (log scale)")
plt.title(
    "Distribution of mean overlap count per Image\n(S-NPP, JPSS-1, JPSS-2)"
)

plt.grid(which="both", alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
print(mean_count_per_dt.min())
print(mean_count_per_dt.max())
print(mean_count_per_dt.mean())
print(mean_count_per_dt.std())
```
